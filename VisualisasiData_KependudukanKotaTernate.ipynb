```python
# Rename columns for clarity
df = df.rename(columns={'Unnamed: 0': 'Kelompok Umur', '2015': 'Laki-laki', '2015.1': 'Perempuan'})

# Remove the last row which contains totals
df = df.drop(df.index[-1])

# Convert population columns to numeric (they appear to be int64 now, but this is a safe step)
df['Laki-laki'] = pd.to_numeric(df['Laki-laki'], errors='coerce')
df['Perempuan'] = pd.to_numeric(df['Perempuan'], errors='coerce')

# Invert the values for the male population to create the pyramid shape
df['Laki-laki'] = df['Laki-laki'] * -1

# Display the cleaned data
display(df.head())
df.info()
```

**Reasoning**:
Create the population pyramid using the cleaned data.

```python
import matplotlib.pyplot as plt

# Create the population pyramid
fig, axes = plt.subplots(figsize=(12, 8), ncols=2, sharey=True)

# Plot male population on the left
axes[0].barh(df['Kelompok Umur'], df['Laki-laki'], align='center', color='skyblue')
axes[0].set(title='Laki-laki')

# Plot female population on the right
axes[1].barh(df['Kelompok Umur'], df['Perempuan'], align='center', color='salmon')
axes[1].set(title='Perempuan')

# Set common labels and title
axes[0].set(yticks=df['Kelompok Umur'], yticklabels=df['Kelompok Umur'])
axes[0].invert_xaxis()
axes[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: f'{abs(x):,.0f}'))
axes[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: f'{abs(x):,.0f}'))
axes[1].autoscale(enable=True, axis='x', tight=True)


fig.suptitle('Piramida Penduduk Kota Ternate, 2015', y=1.02)
plt.tight_layout()
plt.show()
```
