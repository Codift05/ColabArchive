```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset yang sudah kita susun sebelumnya (fitur utama)
data = {
    "Tahun": [2017, 2018, 2019, 2020, 2021, 2022],
    "Produksi": [49481, 5800, 11837, 9291, 4576, 4812],
    "Harga Lokal": [29615, 23276, 25109, 24768, 25951, 26486],
    "Harga Internasional (USD/Kg)": [3.0, 3.1, 3.3, 3.4, 3.2, 3.5],
    "Kurs": [13500, 14200, 14100, 14300, 14200, 14500],
    "Musim": ["Kemarau", "Hujan", "Kemarau", "Hujan", "Hujan", "Kemarau"],
    "Frekuensi Penangkapan": [20, 16, 18, 17, 15, 18],
    "Konsumsi": [4000, 3600, 3700, 3550, 3500, 3800],
    "Alat Tangkap": ["Purse Seine", "Handline", "Gillnet & Handline", "Gillnet & Handline", "Handline (Tradisional)", "Purse Seine"]
}

df = pd.DataFrame(data)

# Konversi harga internasional ke Rupiah
df["Harga Internasional"] = df["Harga Internasional (USD/Kg)"] * df["Kurs"]

# Hapus kolom yang tidak dibutuhkan langsung
df.drop(["Harga Internasional (USD/Kg)", "Kurs"], axis=1, inplace=True)

# Tampilkan data
df.head()
```

```python
# Encode Musim dan Alat Tangkap
label_encoder_musim = LabelEncoder()
label_encoder_alat = LabelEncoder()

df["Musim (Encoded)"] = label_encoder_musim.fit_transform(df["Musim"])
df["Alat Tangkap (Encoded)"] = label_encoder_alat.fit_transform(df["Alat Tangkap"])

# Drop kolom kategorikal asli (opsional)
df_model = df.drop(["Tahun", "Musim", "Alat Tangkap"], axis=1)

df_model.head()
```

```python
# Normalisasi menggunakan MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_model)

# Konversi ke DataFrame lagi
df_scaled = pd.DataFrame(scaled_data, columns=df_model.columns)
df_scaled.head()
```

```python
# Elbow Method dengan range yang sesuai
wcss = []
K_range = range(1, 7)  # Karena jumlah data hanya 6

for k in K_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(10,6))
plt.plot(K_range, wcss, marker='o')
plt.title('Metode Elbow untuk Menentukan Jumlah Cluster Optimal')
plt.xlabel('Jumlah Cluster (K)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
# K-Means Clustering dengan K=2
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(df_scaled)

# Tambahkan hasil cluster ke dataframe asli
df['Cluster'] = y_kmeans

# Lihat hasil
print(df[['Tahun', 'Produksi', 'Harga Lokal', 'Harga Internasional', 'Frekuensi Penangkapan', 'Musim', 'Konsumsi', 'Alat Tangkap', 'Cluster']])
```

```python
# Visualisasi Hasil Clustering: Produksi vs Harga Lokal
plt.figure(figsize=(10,6))

# Plot Cluster 0
sns.scatterplot(
    x=df['Produksi'],
    y=df['Harga Lokal'],
    hue=df['Cluster'],
    palette='coolwarm',
    s=100
)

plt.title('Visualisasi Cluster: Produksi vs Harga Lokal')
plt.xlabel('Produksi (Ton)')
plt.ylabel('Harga Lokal (Rp/Kg)')
plt.grid(True)
plt.legend(title='Cluster')
plt.show()
```
